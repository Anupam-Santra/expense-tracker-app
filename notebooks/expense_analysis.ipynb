## 1. Setup & Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sys, os
sys.path.insert(0, '..')
warnings.filterwarnings('ignore')

# Style
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['axes.facecolor'] = '#F8F9FA'
plt.rcParams['figure.facecolor'] = '#F8F9FA'
sns.set_palette('husl')

print('✅ Libraries loaded')
```

## 2. Generate & Load Data

```python
from src.data_generator import generate_expenses
from src.data_cleaner import clean_data

df_raw = generate_expenses(year=2024, num_records=500)
df = clean_data(df_raw)

print(f'Shape: {df.shape}')
print(f'Columns: {list(df.columns)}')
df.head(10)
```

## 3. Dataset Overview

```python
print('='*50)
print('DATASET INFO')
print('='*50)
print(f'Total Records      : {len(df)}')
print(f'Date Range         : {df["Date"].min().date()} → {df["Date"].max().date()}')
print(f'Total Spend        : ₹{df["Amount"].sum():,.2f}')
print(f'Avg Monthly Spend  : ₹{df["Amount"].sum()/12:,.2f}')
print(f'Avg Transaction    : ₹{df["Amount"].mean():,.2f}')
print(f'Max Transaction    : ₹{df["Amount"].max():,.2f}')
print(f'Min Transaction    : ₹{df["Amount"].min():,.2f}')
print(f'Categories         : {df["Category"].nunique()}')
print(f'Payment Methods    : {df["Payment_Method"].nunique()}')
df.describe()
```

## 4. Missing Values & Data Quality

```python
print('Missing Values:')
print(df.isnull().sum())
print(f'\nData Types:')
print(df.dtypes)
```

## 5. Category-wise Analysis

```python
from src.analysis import category_summary
cat = category_summary(df)
print(cat.to_string(index=False))

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Pie
axes[0].pie(cat['Total'], labels=cat['Category'], autopct='%1.1f%%', startangle=140,
            wedgeprops={'linewidth':1.5,'edgecolor':'white'})
axes[0].set_title('Category-wise Spending Distribution', fontweight='bold')

# Bar
cat_s = cat.sort_values('Total')
axes[1].barh(cat_s['Category'], cat_s['Total'], edgecolor='white')
axes[1].set_title('Total Spend by Category', fontweight='bold')
axes[1].set_xlabel('Amount (₹)')

plt.tight_layout()
plt.show()
```

## 6. Monthly Trend Analysis

```python
from src.analysis import monthly_trend
monthly = monthly_trend(df)
print(monthly.to_string(index=False))

fig, ax = plt.subplots(figsize=(14, 5))
ax.fill_between(monthly['Month'], monthly['Total'], alpha=0.15)
ax.plot(monthly['Month'], monthly['Total'], marker='o', linewidth=2.5, label='Total Spend')
ax.set_title('Monthly Spending Trend - 2024', fontweight='bold', fontsize=14)
ax.set_ylabel('Total Spend (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 7. Payment Method Analysis

```python
from src.analysis import payment_method_analysis
pm = payment_method_analysis(df)
print(pm.to_string(index=False))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].bar(pm['Payment_Method'], pm['Total'], edgecolor='white')
axes[0].set_title('Spend by Payment Method')
axes[0].tick_params(axis='x', rotation=20)

axes[1].pie(pm['Count'], labels=pm['Payment_Method'], autopct='%1.0f%%', startangle=90,
            wedgeprops={'linewidth':1.5,'edgecolor':'white'})
axes[1].set_title('Transaction Count by Payment Method')
plt.tight_layout()
plt.show()
```

## 8. Heatmap - Month × Category

```python
pivot = df.pivot_table(index='Month_Num', columns='Category', values='Amount',
                       aggfunc='sum', observed=True).fillna(0)
pivot.index = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec'][:len(pivot)]

fig, ax = plt.subplots(figsize=(14, 5))
sns.heatmap(pivot, ax=ax, cmap='YlOrRd', linewidths=0.5, annot=True, fmt='.0f',
            annot_kws={'size':7})
ax.set_title('Spending Heatmap: Month × Category', fontweight='bold', fontsize=14)
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

## 9. Budget vs Actual

```python
from src.analysis import budget_vs_actual
bva = budget_vs_actual(df)
print(bva.to_string(index=False))

x = range(len(bva))
fig, ax = plt.subplots(figsize=(12, 5))
ax.bar([i-0.2 for i in x], bva['Budget'],     0.38, label='Budget',  alpha=0.85)
ax.bar([i+0.2 for i in x], bva['Actual_Avg'], 0.38, label='Actual',  alpha=0.85)
ax.set_xticks(list(x))
ax.set_xticklabels(bva['Category'], rotation=30, ha='right')
ax.set_title('Monthly Budget vs Actual Spending', fontweight='bold', fontsize=14)
ax.legend()
plt.tight_layout()
plt.show()
```
